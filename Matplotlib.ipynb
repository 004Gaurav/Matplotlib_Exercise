```python
import numpy as np

x = ["Python","c","c++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Graph",fontsize=15)

plt.bar(p,y,width, color="r", label="popularity")
plt.bar(p1,z,width, color="y", label="popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]

plt.scatter(day,no)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
day = [1, 2, 3, 4, 5, 6, 7]  # X-axis values (days)
no = [2, 3, 1, 4, 5, 3, 6]   # Y-axis values (numbers)
colors = ["r", "y", "b", "g", "y", "r", "g"]  # Colors for each point
sizes = [100, 200, 400, 500, 200, 100, 700]   # Sizes of the points

# Creating the scatter plot
plt.scatter(day, no, c=colors, s=sizes,alpha =0.4,marker="*",linewidth=2)

# Adding title and labels
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
day = [1, 2, 3, 4, 5, 6, 7]  # X-axis values (days)
no = [2, 3, 1, 4, 5, 3, 6]   # Y-axis values (numbers)
colors = [100, 200, 400, 500, 200, 100, 700]  # Colors for each point
sizes = [100, 200, 400, 500, 200, 100, 700]   # Sizes of the points

# Creating the scatter plot with a color map
plt.scatter(day, no, c=colors, s=sizes, cmap="viridis")  # Plotting the scatter points
color_bar = plt.colorbar()  # Adding a color bar to indicate value-color mapping
color_bar.set_label("ColorBar")  # Labeling the color bar

# Adding title and labels
plt.title("Scatter Plot", fontsize=15)  # Setting the title of the plot
plt.xlabel("Day", fontsize=15)  # Labeling the x-axis
plt.ylabel("Number", fontsize=15)  # Labeling the y-axis

# Display the plot
plt.show()  # Displaying the scatter plot with all its elements
```

# Pie Plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","Java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,
        radius = 1.5,labeldistance=0.8,startangle=180,
        textprops={"fontsize": 15},counterclock=False,
        wedgeprops={"linewidth":5})

plt.legend(loc=2)
plt.show()
```

# Area vs Stack Plot

```python
import matplotlib.pyplot as plt

x     = [1, 2, 3, 4, 5]

area1 = [2, 3, 2, 5, 4]
area2 = [4, 7, 9, 3, 8]
area3 = [6, 2, 5,10, 1]

l= ["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","y"])

plt.legend(loc=2)
plt.show
```

# Fil_Between Plot

```python
import matplotlib.pyplot as plt

x= [1,2,3,4,5]
y= [1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array ([1,2,3,4,5])
y=np.array ([1,2,3,4,5])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4))

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
