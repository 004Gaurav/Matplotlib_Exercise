```python
# Import the necessary library
import matplotlib.pyplot as plt

# Data for the scatter plot
day = [1, 2, 3, 4, 5, 6, 7]  # X-axis values (days)
no = [2, 3, 1, 4, 5, 3, 6]   # Y-axis values (numbers)
colors = ["r", "y", "b", "g", "y", "r", "g"]  # Colors for each point
sizes = [100, 200, 400, 500, 200, 100, 700]   # Sizes of the points

# Creating the scatter plot using plt.scatter()
plt.scatter(day, no, c=colors, s=sizes, alpha=0.4, marker="*", linewidth=2)

# Adding title and labels to the plot
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)

# Display the plot
plt.show()
```

# Pie Plot

```python
import matplotlib.pyplot as plt

# Data for the pie chart
x = [10, 20, 30, 40]  # Values for the pie chart
y = ["c", "c++", "Java", "python"]  # Labels for each slice
ex = [0.4, 0.0, 0.0, 0.0]  # List of explode values for each slice
c = ["r", "b", "g", "y"]  # List of colors for each slice

# Creating the pie chart with various customizations
plt.pie(x,  # Values for the slices
        labels=y,  # Labels for each slice
        explode=ex,  # Amount of separation between slices
        autopct="%0.1f%%",  # Format for displaying the percentage on each slice
        shadow=True,  # Add shadow to the chart
        radius=1.5,  # Radius of the pie chart
        labeldistance=0.8,  # Distance of labels from the center of the pie chart
        startangle=180,  # Starting angle of the first slice (180 degrees = top of the circle)
        textprops={"fontsize": 15},  # Font size of the labels and percentages
        counterclock=False,  # Direction of the slices (clockwise)
        wedgeprops={"linewidth": 5}  # Width of the border lines for each slice
        )

# Adding legend to the pie chart
plt.legend(loc=2)  # Location of the legend (2 = upper left corner)

# Display the pie chart
plt.show()  # Displaying the pie chart with all its customizations
```

In this code:

1. We define the `x` list containing the values for each slice, the `y` list containing labels for each slice, the `ex` list containing explode values to separate slices, and the `c` list containing colors for each slice.
2. We use `plt.pie()` to create the pie chart with various customizations:
   - `explode` separates slices by the specified amount.
   - `autopct` formats and displays the percentage on each slice.
   - `shadow` adds shadow to the chart.
   - `radius` sets the radius of the pie chart.
   - `labeldistance` adjusts the distance of labels from the center.
   - `startangle` sets the starting angle of the first slice.
   - `textprops` sets the font size of labels and percentages.
   - `counterclock` sets the direction of the slices (clockwise).
   - `wedgeprops` sets the width of border lines for each slice.
3. We add a legend to the pie chart using `plt.legend()` and specify the location of the legend.
4. Finally, we use `plt.show()` to display the customized pie chart.

```python
import matplotlib.pyplot as plt

# Data for the pie chart layers
x = [10, 20, 30, 40]  # Values for the outer layer
x1 = [40, 30, 20, 10]  # Values for the inner layer
y = ["c", "c++", "Java", "python"]  # Labels for each slice
c = ["r", "b", "g", "y"]  # List of colors for each slice

# Creating the double-layered pie chart
plt.pie(x, labels=y, radius=1.5)  # Outer layer
plt.pie(x1, colors="w")  # Inner layer with a single slice to create a transparent center

# Display the pie chart
plt.show()  # Displaying the double-layered pie chart
```

In this code:

1. We define the `x` list containing the values for the outer layer of the pie chart, the `x1` list containing the values for the inner layer, the `y` list containing labels for each slice, and the `c` list containing colors for each slice.
2. We use `plt.pie()` twice to create the double-layered pie chart:
   - The first `plt.pie()` call creates the outer layer using the `x` values and `y` labels.
   - The second `plt.pie()` call creates the inner layer with a single slice using the `x1` values. We use the `colors="w"` parameter to make this slice transparent, creating a hole in the center of the pie chart.
3. Finally, we use `plt.show()` to display the double-layered pie chart with the transparent center.

```python
import matplotlib.pyplot as plt

# Data for the pie chart
x = [10, 20, 30, 40]  # Values for the slices
y = ["c", "c++", "Java", "python"]  # Labels for each slice
c = ["r", "b", "g", "y"]  # List of colors for each slice

# Creating the pie chart
plt.pie(x, labels=y, radius=1.5)  # Outer layer

# Adding a transparent center circle to create a hole
center_circle = plt.Circle(xy=(0, 0), radius=1, facecolor="w")  # Create a circle at the center with white color
plt.gca().add_artist(center_circle)  # Add the circle to the current axes

# Display the pie chart
plt.show()  # Displaying the pie chart with the transparent center
```

In this code:

1. We define the `x` list containing the values for the slices, the `y` list containing labels for each slice, and the `c` list containing colors for each slice.
2. We use `plt.pie()` to create the pie chart with the specified values and labels.
3. We create a transparent center circle using `plt.Circle()`. The `xy` parameter sets the center coordinates of the circle, the `radius` parameter defines the radius of the circle, and the `facecolor` parameter sets the color of the circle (in this case, "w" for white, which makes it transparent).
4. We add the transparent center circle to the current axes using `plt.gca().add_artist(center_circle)`.
5. Finally, we use `plt.show()` to display the pie chart with the transparent center circle, creating a hole in the center of the chart.

# Stem Plot

In this code:

2. We use `plt.stem()` to create the stem plot. The `linefmt` parameter sets the format for the stem lines, the `markerfmt` parameter sets the format for the markers at the top of the stems, the `bottom` parameter specifies the y-coordinate for the baseline of the stems, the `basefmt` parameter sets the format for the baseline, the `label` parameter assigns a label to the data series, and the `use_line_collection` parameter is set to `False` to use a LineCollection instead of a PathCollection (you can omit this parameter if you're using a version of Matplotlib where LineCollection is the default).
3. We add a legend to the plot using `plt.legend()`.
4. Finally, we use `plt.show()` to display the stem plot with the customized formatting and legend.

# Box Plot

```python
import matplotlib.pyplot as plt

# Data for the box plot
x = [10, 20, 30, 40, 50, 60, 70]

# Creating the box plot
plt.boxplot(x, patch_artist=True, showmeans=True)

# Display the plot
plt.show()

```

In this code:

1. We define the `x` list containing the data values for the box plot.
2. We use `plt.boxplot()` to create the box plot. The `patch_artist` parameter is set to `True` to enable filling of the boxes with color, and the `showmeans` parameter is set to `True` to display the mean value as a marker.
3. Finally, we use `plt.show()` to display the box plot with the customized formatting and features. The boxes are filled with color, and the mean value is shown as a marker.

```python
import matplotlib.pyplot as plt

# Data for the box plot
x = [10, 20, 30, 40, 50, 60, 70, 120]

# Creating the box plot
plt.boxplot(x, showmeans=True, sym="g*")

# Display the plot
plt.show()
```

In this code:

1. We define the `x` list containing the data values for the box plot.

# Fil_Between Plot

```python
import matplotlib.pyplot as plt

# Data for the line plot
x = [1, 2, 3, 4, 5]  # X-axis values
y = [1, 2, 3, 4, 5]  # Y-axis values

# Creating the line plot
plt.plot(x, y, color="r")  # Plotting the line

# Filling the region between two horizontal lines
plt.fill_between(x=[2, 4], y1=2, y2=4)  # Filling the area between y=2 and y=4

# Adding title and labels to the axes
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Display the plot
plt.show()
```

Here's what each part of the code does:

1. You define the x-axis values `x` and y-axis values `y` for the line plot.
2. You use `plt.plot()` to create the line plot with the specified data points and color (`color="r"` for red).
3. You use `plt.fill_between()` to fill the region between two horizontal lines:
   - `x=[2, 4]` specifies the range of x values for the filled region.
   - `y1=2` and `y2=4` specify the y values for the bottom and top of the filled region.
4. You add a title to the plot using `plt.title()`.
5. You add labels to the x and y axes using `plt.xlabel()` and `plt.ylabel()`.
6. Finally, you use `plt.show()` to display the line plot with the shaded region and the specified title and labels.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the line plot
x = np.array([1, 2, 3, 4, 5])  # X-axis values
y = np.array([1, 2, 3, 4, 5])  # Y-axis values

# Creating the line plot
plt.plot(x, y, color="r")  # Plotting the line

# Filling the region between two vertical lines
plt.fill_between(x, y, color="g", where=(x >= 2) & (x <= 4))  # Filling the area between x=2 and x=4

# Adding title and labels to the axes
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Display the plot
plt.show()
```

Here's what each part of the code does:

1. You import the necessary libraries: `matplotlib.pyplot` as `plt` and `numpy` as `np`.
2. You define the x-axis values `x` and y-axis values `y` for the line plot as NumPy arrays.
3. You use `plt.plot()` to create the line plot with the specified data points and color (`color="r"` for red).
4. You use `plt.fill_between()` to fill the region between two vertical lines:
   - `x` is the x values for the filled region.
   - `y` is the y values for the filled region (in this case, it's the same as the y values of the line plot).
   - `color="g"` specifies the color of the shaded region.
   - `where=(x >= 2) & (x <= 4)` defines the condition for filling the region between x=2 and x=4.
5. You add a title to the plot using `plt.title()`.
6. You add labels to the x and y axes using `plt.xlabel()` and `plt.ylabel()`.
7. Finally, you use `plt.show()` to display the line plot with the shaded region and the specified title and labels.
